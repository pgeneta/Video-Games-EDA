<h1>What is the most popular gaming genre?</h1>

```python

display(df['Genre'].value_counts())

genre_index = df['Genre'].value_counts().index
```

```python
sns.countplot(data = df, x = 'Genre', order=genre_index)
plt.xticks(rotation = 45)
plt.title("Most Popular Gaming Genre")
```

<h1>Which year had the most popular gaming release?</h1>

```python
pop_index = df.groupby('Year')['Name'].count().sort_values(ascending=False).index
df.groupby('Year')['Name'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(14,8))
sns.countplot(data=df, x='Year', order=pop_index)
plt.xticks(rotation = 90)
plt.title("Most popular gaming release by year")
```

<h1>Which year had the highest sales by year?</h1>

```python
global_df = df.groupby('Year')['Global_Sales'].sum().reset_index()
display(global_df)
plt.figure(figsize=(14,8))
sns.barplot(data=global_df, x='Year', y='Global_Sales')
plt.xticks(rotation=45)
```

<h1>Sales comparison by genre</h1>

```python
comparison = df[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].groupby('Genre').sum()


comparison
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(data=comparison, annot=True, fmt='.1f')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

```python
comp_table = comparison.reset_index()
comp_table = pd.melt(comp_table, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='sale_area', value_name='sale_price')

comp_table.head()
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=comp_table, x= 'Genre', y = 'sale_price', hue = 'sale_area')
plt.ylabel("Sale Price")
plt.legend(title = "Sale Area", loc = 'upper right')
```

<h1>Sales comparison by platofrm</h1>

```python
comp_platform = df[['Platform','NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_platform = comp_platform.groupby('Platform').sum().reset_index()

comp_platform
```

```python
comp_platform_melt = comp_platform.melt(id_vars='Platform', value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', "Other_Sales"], var_name= 'region', value_name='price')
comp_platform_melt
```

```python
plt.figure(figsize=(30,15))
sns.barplot(data = comp_platform_melt, x = 'Platform', y = 'price', hue='region')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.ylabel('Sale Price')
```

Playstation, DS and XBOX 360 are market leaders
